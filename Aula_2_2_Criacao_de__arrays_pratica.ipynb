```python
#criação de array 1D: [1,2,3]
#l = [1,2,3]
x = np.array([1,2,3])
print('x = ', x)
print('o shape é:', x.shape)
```

```python
#criação de array 1D: [1,2,3] e forçando ser do tipo int
l = [1,2,3]
x = np.array(l, dtype=int)
print('x = ', x)
print('o shape é:', x.shape)
```

```python
#criação de array 2D: [1,2], [3,4]
l = ([1,2], [3,4])
x = np.array(l)
print('x:\n', x)
type(x)
print('o shape é:', x.shape)
```
